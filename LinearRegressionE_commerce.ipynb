```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/Ecommerce'
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Exploratory data analysis

explore the correlation between the yearly amount spent and time on website and application


```python
sns.lmplot(x='Time on Website', y='Yearly Amount Spent', data= df, scatter_kws={'alpha':0.5})
#There is no correlation between ('Time on Website', 'Yearly Amount Spent')
```

```python
sns.lmplot(x='Time on App', y='Yearly Amount Spent', data= df, scatter_kws={'alpha':0.5})
# there is a positive correlation between ('Time on App', 'Yearly Amount Spent')

```

#Split the data using Scikit-Learn

```python
!pip install scikit-learn
from sklearn.model_selection import train_test_split
```

```python
#predictors
x= df[['Avg. Session Length','Time on App', 'Time on Website', 'Length of Membership']]
#target
y= df['Yearly Amount Spent']

```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

#The model

```python
from sklearn.linear_model import LinearRegression
lm= LinearRegression()
```

```python
#fitting the data into the model
lm.fit(x_train, y_train)
```

```python
#interpreting the coefficients
cdf = pd.DataFrame(lm.coef_, x.columns, columns=['Coeff'])
cdf
```

```python
#predictions
predictions = lm.predict(x_test)
predictions
```

```python
#graphic evaluation of predictions
sns.scatterplot(x=predictions, y=y_test)
```

```python
#analysis evaluation of errors
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math

```

```python
print ('MAE:', mean_absolute_error(y_test, predictions))
print ('MSE:', mean_squared_error(y_test, predictions))
print ('RMSE:', math.sqrt(mean_squared_error(y_test, predictions)))
```

```python
#residuals
residuals = y_test + predictions
sns.distplot(residuals)
```

```python
import pylab
import scipy.stats as stats
stats.probplot(residuals, dist="norm", plot=pylab)
pylab.show()
```
